```python
tb = pd.pivot_table(bb97, values="Song", index=["WeekID","Performer"],aggfunc=np.count_nonzero)
```

```python
from collections import defaultdict
cnt = {}
pfcnt = {}

for i in range(1959, 2018):
    cnt[i] = defaultdict(int)
    pfcnt[i] = set()
    
pd = bb
week_performer = pd[['WeekID','Performer']]

pfset = set()
for i, r in week_performer.iterrows():
#     print(i, r["Performer"])
    year = r['WeekID'].split("-")[0]
    if type(r['Performer']) != str:
        print("missing record:", r)
        continue
    pfs = r['Performer'].split("Featuring") # split on "," "&"
#     print(year)
#     print(pfs)
    for p in pfs:
        cnt[int(year)][p.strip()]+=1
        pfcnt[int(year)].add(p)
        pfset.add(p)
```

```python
from collections import defaultdict
import re
missingdata = 0
songCnt = {}
beforeSplit = {}
afterSplit = {}
performers = set()

for i in range(1959, 2018):
    songCnt[i] = defaultdict(int)
    beforeSplit[i] = set()
    afterSplit[i] = set()
    
pd = bb
week_performer = pd[['WeekID','Performer']]


for i, r in week_performer.iterrows():
#     print(i, r["Performer"])
    year = r['WeekID'].split("-")[0]
    if type(r['Performer']) != str:
        print("missing record:", r)
        missingdata = missingdata + 1
        continue
#     re.split("")
    beforeSplit[int(year)].add(r['Performer'])
    pfs = re.split(r"(Featuring|,)",r['Performer'])
#     pfs = r['Performer']replace("").split("Featuring") # split on "," "&"
#     print(year)
#     print(pfs)
    for p in pfs:
        songCnt[int(year)][p.strip()]+=1
        afterSplit[int(year)].add(p)
        performers.add(p)

```

```python
import pandas as pd
from collections import defaultdict
from seaborn import lineplot
year_singer = defaultdict(int)
cleaned_year_singer = defaultdict(int)

for y, pfs in beforeSplit.items():
    year_singer[y] = len(pfs)
for y, pfs in afterSplit.items():
    cleaned_year_singer[y] = len(pfs)
    
ys = pd.DataFrame(list(year_singer.items()),columns=["year","number of popular singers"])
ays = pd.DataFrame(list(cleaned_year_singer.items()),columns=["year","number of popular singers"])
ys["Cleaning"]="Before"
ays["Cleaning"] = "After"
singers = pd.concat([ys, ays])
```
